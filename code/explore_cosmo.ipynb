```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dtype=[('BoxID', 'i8'), ('pairs_i', 'i8'), ('halos_i', 'i8'), 
        ('pairs_f', 'i8'), ('halos_f', 'i8'), ('mean_vel', 'f8'), ('std_vel','f8')]
pair_data = np.loadtxt("../data/summary_pair_count.dat", dtype=dtype)

dtype=[('BoxID','i8'), ('hubble', 'f8'), ('omega_de', 'f8'),
      ('omega_m', 'f8'), ('n_s', 'f8'), ('sigma_8', 'f8'), ('w_0', 'f8')]
cosmo_data = np.loadtxt("../data/box_cosmo_params.dat", dtype=dtype)
```

```python
ii = np.in1d(cosmo_data['BoxID'],pair_data['BoxID'])
print(ii)

#r = pair_data['pair_f']/1000
r = pair_data['mean_vel']
#r = pair_data['pairs_f']/pair_data['pairs_i']

plt.figure(figsize=(18,4))

keys = ['hubble', 'omega_de', 'omega_m', 'n_s', 'sigma_8', 'w_0']

for k, kk in enumerate(keys):
    print(k, kk)
    plt.subplot(1,6,k+1)
    plt.scatter(cosmo_data[kk][ii], r)
    plt.ylim(np.min(r),np.max(r))
    plt.xlabel(kk)
```
